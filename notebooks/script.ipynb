```python
import requests
import pandas as pd
```

funcao para fazer o request na api

Estou utilizando a porta 5001 pois ja havia um processo rodando na 5000

```python
def get_paises():
    response = requests.get('http://127.0.0.1:5001/country')
    return response.json()
```

Fazer a leitura e assignment desses dados

```python

paises_data = get_paises()
paises_df = pd.DataFrame.from_dict(paises_data, orient='index')
paises_df.reset_index(inplace=True)
paises_df.columns = ['codigo', 'nome', 'moeda', 'populacao', 'capital', 'continente', 'area']
```

Visualizar os dados

```python
paises_df
```

Ler os dados do CSV

```python
gdp_df = pd.read_csv('../data/gdp_data.csv')
gdp_df.head()
```

Ajustar nomes das colunas para garantir que correspondam

```python
gdp_df = gdp_df.rename(columns={'Country Code': 'codigo'})
```

Mesclar os dados com base no código do país

```python
merged_df = pd.merge(paises_df, gdp_df, on='codigo', how='left')
```

Vizualizar todas as colunas e verificar se está tudo certinho

```python
print(merged_df.columns)
```

Identifiquei uma coluna chamada 'Unamed: 66', pode ter sido adicionada durante o processo de merge, como não iremos utiliza-la, irei fazer um drop

```python
merged_df.drop(columns=['Unnamed: 66'], inplace=True)
```

Dando un check na versao final antes de exportar para um csv
